# Usecase 3
Display top 5 most demanded products from CSV_s1

```python
products["productno"] = products["productno"].apply(lambda x: x.replace("Product", ""))
products
```

```python
products["productno"] = products["productno"].astype('int')
products
```

```python
demanding01 = pd.merge(products,transactions, on='productno', how='inner')
demanding01
```

```python
demanding01
```

```python
demanding02 = demanding01.groupby(['productName'])['productno'].count().reset_index(
  name='Quantity Sold').sort_values(['Quantity Sold'], ascending=False)
```
